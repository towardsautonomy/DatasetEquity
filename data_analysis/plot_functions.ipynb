```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dequity_loss(eta: float, p: float, gamma: float) -> float:
    return (eta + (1 - p) ** gamma) / (eta + 1)
```

```python
# plot functions for various values of eta and gamma
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xlabel('Sample Likelihood', fontsize=16)
ax.set_ylabel('Dequity Loss Weight', fontsize=16)
ax.set_title('Generalized Focal Loss', fontsize=20)
ax.grid(alpha=0.5, linestyle='--', linewidth=1)

eta = 0.0
p = np.linspace(0, 1, 100)
gamma = 2
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 0.0, $\gamma$ = 2', linewidth=4)

eta = 0.0
p = np.linspace(0, 1, 100)
gamma = 5
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 0.0, $\gamma$ = 5', linewidth=4)

eta = 0.3
p = np.linspace(0, 1, 100)
gamma = 2
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 0.3, $\gamma$ = 2', linewidth=4)

eta = 0.3
p = np.linspace(0, 1, 100)
gamma = 5
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 0.3, $\gamma$ = 5', linewidth=4)

eta = 1.0
p = np.linspace(0, 1, 100)
gamma = 2
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 1.0 $\gamma$ = 2', linewidth=4)

eta = 1.0
p = np.linspace(0, 1, 100)
gamma = 5
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 1.0, $\gamma$ = 5', linewidth=4)

eta = 3
p = np.linspace(0, 1, 100)
gamma = 2
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 3.0, $\gamma$ = 2', linewidth=4)

eta = 3
p = np.linspace(0, 1, 100)
gamma = 5
ax.plot(p, dequity_loss(eta, p, gamma), label=r'$\eta$ = 3.0, $\gamma$ = 5', linewidth=4)

# eta = 0
# p = np.linspace(0, 1, 100)
# gamma = 1
# ax.plot(p, dequity_loss(eta, p, gamma), label='eta = 0.0, gamma = 1')

# fancy legend
plt.legend(loc='upper right', fancybox=True, framealpha=0.8, fontsize=14, 
           ncol=2, facecolor='white', edgecolor='black', shadow=True)
# shadow
ax.spines['top'].set_linewidth(0.5)
ax.spines['right'].set_linewidth(0.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
plt.savefig('plots/dequity_loss.png', dpi=300, bbox_inches='tight')
plt.show()
```
